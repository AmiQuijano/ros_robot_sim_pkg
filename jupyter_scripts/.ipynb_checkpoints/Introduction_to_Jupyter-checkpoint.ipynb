## Matplotlib

````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
# Usually imports are in one separate cell
````

````python
%matplotlib inline
x = np.arange(0, math.pi*2, 0.05)
y= np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title("sine_wave")
ax.set_xlabel("angle")
ax.set_ylabel("sin")
````

````python
%matplotlib widget
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title("sine_wave")
ax.set_xlabel("angle")
ax.set_ylabel("sin")
````

````python
# We can add another plot in the same graph
y = np.sin(2*x)
ax.plot(x,y)
````
